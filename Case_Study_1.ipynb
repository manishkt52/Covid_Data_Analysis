**Q2.2: Generate plots of confirmed cases over time for the top countries.**

```python
data_covid_confirm.drop(columns=['Lat', 'Long'], inplace=True)
data_covid_confirm.head()
```

```python
list_col = [];
for col in data_covid_confirm.columns:
  if data_covid_confirm[col].dtype == 'int64':
    list_col.append(col)

print(list_col)
```

```python
value = data_covid_confirm.groupby('Country/Region')[list_col].agg('sum')
value.head()
```

```python
value_transposed = value.T
value_transposed.head()
```

```python
top_countries = value.sum(axis=1).sort_values(ascending=False).head(5).index
value_transposed[top_countries].plot(figsize=(12,8))

plt.title('Confirmed COVID-19 Cases Over Time (Top 5 Countries)')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Country')
plt.show()
```

**Q2.3: Generate plots of confirmed cases over time for China.**

```python
china_covid_data = value.loc['China']
china_covid_data
```

```python
updated_data_covid = china_covid_data.T
updated_data_covid.head()
```

```python
updated_data_covid.plot(figsize=(10, 5), color='red')

plt.title('COVID-19 Confirmed Cases Over Time in China')
plt.xlabel('Date')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
